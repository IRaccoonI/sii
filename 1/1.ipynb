```python
from enum import Enum


df.rename(columns={
    col_name: col_name.replace(' ', '_').replace('(', '_').replace(')', '').lower() for col_name in df.columns
}, inplace=True)

class DfCols:
    country = 'country' 
    region = 'region' 
    happiness_rank = 'happiness_rank' 
    happiness_score = 'happiness_score'
    standard_error = 'standard_error' 
    economy__gdp_per_capita = 'economy__gdp_per_capita' 
    family = 'family'
    health__life_expectancy = 'health__life_expectancy' 
    freedom = 'freedom' 
    trust__government_corruption = 'trust__government_corruption'
    generosity = 'generosity' 
    dystopia_residual = 'dystopia_residual'
    sum = 'sum'
    happiness_rounded = 'happiness_rounded'

df[DfCols.sum] = 0
df[DfCols.happiness_rounded] = 0

df.columns
```
